```python
import pandas as pd
import json
```

```python
with open("../world_airports.json", "r", encoding="utf-8") as f:
    flights_data = json.load(f)
```

```python
df = pd.DataFrame(flights_data)
df.head()
```

```python
print(df.columns)
```

```python
df['geom'] = 'SRID=4326;POINT(' + df['longitude_deg'].astype(str) + ' ' + df['latitude_deg'].astype(str) + ')'
df.head()
```

```python
df = df.drop(columns=['latitude_deg', 'longitude_deg'])

df.head()
```

```python
print(df.columns)
```

```python
desired_order = [
    "iata_code",
    "icao_code",
    "name",
    "type",
    "geom",
    "continent",
    "country_name",
    "iso_country",
    "region_name",
    "iso_region",
    "municipality",
    "home_link",
    "wikipedia_link"
]
df = df[desired_order]
df.head()
```

```python
# Create a new row as a dictionary
new_row = {col: "" for col in df.columns}
new_row['iata_code'] = "999"
new_row['geom'] = "SRID=4326;POINT(0 0)"
new_row['type'] = "small_airport"
new_row['continent'] = "Africa"

# Append the new row to the DataFrame
df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)

df.tail()
```

```python
df.to_csv('world_airports.txt', sep='\t', index=False, header=False, encoding='utf-8')
```
